```python
import sys
!{sys.executable} -m pip install bar_chart_race

import pandas as pd
df=pd.read_csv('/content/TotalQuestions.csv')
import bar_chart_race as bcr
from IPython.display import Video

# Assuming 'data' is your original DataFrame
df['Date'] = pd.to_datetime(df['Month']).dt.strftime('%Y-%m')
data = df.drop('Month', axis=1)

# Make a copy of the DataFrame
df2 = data.copy()
df2.index = df2['Date'].tolist()
df2 = df2.drop('Date', axis=1)

def make_bcr(df):
    bcr.bar_chart_race(
        df=df,
        filename='stack_overflow.mp4',
        orientation='h',
        sort='desc',
        n_bars=15,
        fixed_order=False,
        fixed_max=False,
        steps_per_period=6,  # Speed control
        interpolate_period=False,
        label_bars=True,
        bar_size=.95,
        period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'},
        period_summary_func=lambda v, r: {'x': .99, 'y': .18, 's': '',
                                          'ha': 'right', 'size': 8, 'family': 'Courier New'},
        period_length=400,
        figsize=(10, 6),  # Larger figure size
        dpi=100,  # Lower DPI for larger size
        cmap='dark24',
        title='Stack Overflow Questions Bar Chart Race',
        title_size=15,  # Larger title size
        bar_label_size=10,  # Larger bar label size
        tick_label_size=8,  # Larger tick label size
        shared_fontdict={'color': '.1'},
        scale='linear',
        writer=None,
        fig=None,
        bar_kwargs={'alpha': .7},
        filter_column_colors=True
    )

# Generate the bar chart race for df2
make_bcr(df2)

# Display the resulting video
Video("stack_overflow.mp4", width=900, height=600, embed=True)
```
